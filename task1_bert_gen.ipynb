```python
# Reading the df
df = pd.read_csv('/kaggle/input/gonordara/train.csv')
test = pd.read_csv('/kaggle/input/gonordara/test.csv') 
sample_submission = pd.read_csv('/kaggle/input/gonordara/sample_submission.csv')
classes_list = sample_submission.columns[1:]
print(len(classes_list))
print(df.shape)
print(test.shape)
print(sample_submission.shape)
df.head()
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load pre-trained DistilRoBERTa model and tokenizer
tokenizer = AutoTokenizer.from_pretrained('nickmuchi/distilroberta-base-movie-genre-prediction')
model = AutoModel.from_pretrained('nickmuchi/distilroberta-base-movie-genre-prediction')

# Move model to the GPU if available
model.to(device)

# Define a function to generate DistilRoBERTa embeddings for a given text
def generate_distilroberta_embeddings(text):
    print("h")
    # Tokenize input text
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
    # Move inputs to the GPU
    inputs = {key: value.to(device) for key, value in inputs.items()}
    # Forward pass through DistilRoBERTa model
    with torch.no_grad():
        outputs = model(**inputs)
    # Move outputs to CPU and extract the hidden states (embeddings) from the DistilRoBERTa model
    embeddings = outputs.last_hidden_state.cpu()
    # Average pooling across the token embeddings
    avg_pooling = torch.mean(embeddings, dim=1)
    return avg_pooling.numpy()

# Generate DistilRoBERTa embeddings for train, validation, and test sets
distilroberta_df = np.vstack([generate_distilroberta_embeddings(text) for text in df['plot_synopsis']])
distilroberta_test = np.vstack([generate_distilroberta_embeddings(text) for text in test['plot_synopsis']])
```

```python
np.save('distilroberta_df.npy', distilroberta_df)
np.save('distilroberta_test.npy', distilroberta_test)
```
